## Admin (Scaling)

```python
import requests
import os

ADMIN_API_URL = "http://triton-inference-server-admin.domino-inference-test.svc.cluster.local:8000/v1/deployments/inference-server/scale"
# Use whichever header your API expects.
# If your Domino instance requires "Authorization: ApiKey <key>" use that.


# --- CALL ENDPOINT --------------------------------------------------

def scale_inference_server(payload,AUTH_HEADER):
    resp = requests.post(
        ADMIN_API_URL,
        headers=AUTH_HEADER,
        json=payload,
        timeout=10
    )

    if resp.status_code != 200:
        print("❌ Error scaling deployment")
        print("Status:", resp.status_code)
        print("Body:", resp.text)
        return

    print("✅ Successfully scaled deployment:")
    print(json.dumps(resp.json(), indent=2))

```

```python
import json
payload = {
    "replicas": 1
}

token = requests.get(os.environ['DOMINO_API_PROXY']+ "/access-token").text

AUTH_HEADER = {
    "Authorization": f"Bearer {token}"
}
scale_inference_server(payload,AUTH_HEADER)
```

## Ask Triton to show all its loaded models

```python
import requests
import os

GET_ALL_MODELS = "http://triton-inference-server-proxy.domino-inference-test.svc.cluster.local:8080/v1/models"
token = requests.get(os.environ['DOMINO_API_PROXY']+ "/access-token").text

AUTH_HEADER = {
    "Authorization": f"Bearer {token}"
}

o = requests.get(
        GET_ALL_MODELS,
        headers=AUTH_HEADER,
        timeout=10)

print(o.json())
```

## Ask Triton to show logs for the Triton deployment

```python
import requests
import os

GET_LOGS = "http://triton-inference-server-admin.domino-inference-test.svc.cluster.local:8000/v1/deployments/inference-server/logs?tail_lines=50"
token = requests.get(os.environ['DOMINO_API_PROXY']+ "/access-token").text

AUTH_HEADER = {
    "Authorization": f"Bearer {token}"
}

o = requests.get(
        GET_LOGS,
        headers=AUTH_HEADER,
        timeout=10)

print(o.json()['logs'])
```
